```python
from tensorflow.keras.datasets import cifar10

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

```python
import matplotlib.pyplot as plt
import cv2

sample_img = X_train[0]

gray = cv2.cvtColor(sample_img, cv2.COLOR_BGR2GRAY)
enhanced = cv2.convertScaleAbs(gray, alpha=1.3, beta=40)

plt.figure(figsize=(9,3))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(sample_img)
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Enhanced")
plt.imshow(enhanced, cmap='gray')
plt.axis('off')

plt.show()


```

```python
import numpy as np

def preprocess_images(images):
    processed = []
    for img in images:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        normalized = gray / 255.0
        processed.append(normalized)
    return np.array(processed)

X_train_p = preprocess_images(X_train)
X_test_p = preprocess_images(X_test)

print("Processed training shape:", X_train_p.shape)
print("Processed testing shape:", X_test_p.shape)
```

```python
X_train_flat = X_train_p.reshape(len(X_train_p), -1)
X_test_flat = X_test_p.reshape(len(X_test_p), -1)

print("Flattened train shape:", X_train_flat.shape)
print("Flattened test shape:", X_test_flat.shape)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train_flat[:5000], y_train[:5000].ravel())

print("Model training completed")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = model.predict(X_test_flat[:1000])

accuracy = accuracy_score(y_test[:1000], y_pred)
print("Model Accuracy:", accuracy)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test[:1000], y_pred))
```

```python
class_names = ['airplane','car','bird','cat','deer',
               'dog','frog','horse','ship','truck']

plt.figure(figsize=(10,6))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(X_test[i])
    plt.title(f"Predicted: {class_names[y_pred[i]]}\nActual: {class_names[y_test[i][0]]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
